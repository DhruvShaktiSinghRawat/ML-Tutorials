```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/homeprices (1).csv')
df.head()
```

```python
df.bedrooms.median()  # we are finding median to replace the value 'NaN' with the median as the value there is missing
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())   # This is done to convert the value of df.bedrooms.median() in integer if it was in decimal
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)  # this has replaced the value of NaN with the value of median_bedrooms
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.predict([[3000, 3, 40]])  # this has predicted the price for a house of 3000 sq. ft, 3 bedrooms & 40 year old
```

```python
data=pd.read_csv('/content/hiring (1).csv')
data.head()
```
